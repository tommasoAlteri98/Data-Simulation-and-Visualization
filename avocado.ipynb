```python
import pandas as pd

# file CSV
file_path = 'avocado.csv'

#dati nel DF
df = pd.read_csv(file_path)

# Lista delle regioni da escludere
regioni_da_escludere = ["TotalUS", "West", "East", "Southweast", "Southeast", "Northeast", "Midsouth", "Plains","SouthCentral", "California"]
# Lista delle regioni ammesse 
regioni_ammissibili = ["East", "West", "Southeast", "North", "Northeast", "Southwest", "Midwest", "SouthCentral"]

# Creazione del nuovo DF mantenendo solo le regioni nella lista
df_raggruppamentiCardinali = df[df['region'].isin(regioni_ammissibili)]
# Creazione del nuovo DF con solo le regioni specifiche degli stati
df_stati = df[~df['region'].isin(regioni_da_escludere)]

#prime righe del DF
print(df.head())
```

```python
#volume totale venduto per anno
volume_totale_anno = df.groupby('year')['Total Volume'].sum()
print(volume_totale_anno)

plt.figure(figsize=(10,6))
sns.lineplot(x=volume_totale_anno.index, y=volume_totale_anno.values)
plt.title("Volume Totale Venduto per Anno")
plt.xlabel("year")
plt.ylabel("Volume Totale")
plt.show()

# Raggruppamento per anno per ciascun PLU
volume_4046 = df.groupby('year')['4046'].sum()
volume_4225 = df.groupby('year')['4225'].sum()
volume_4770 = df.groupby('year')['4770'].sum()

#volume totale di ogni PLU venduto per anno
plt.figure(figsize=(10,6))
sns.lineplot(x=volume_4046.index, y=volume_4046.values, label='4046')
sns.lineplot(x=volume_4225.index, y=volume_4225.values, label='4225')
sns.lineplot(x=volume_4770.index, y=volume_4770.values, label='4770')
plt.title("Volume Venduto per PLU per Anno")
plt.xlabel("year")
plt.ylabel("Volume Venduto")
plt.legend()
plt.show()
```

```python
#volume totale venduto ogni anno
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(10, 3))
sns.histplot(df['year'].dropna(), bins=30, kde=True, color='blue')
plt.xlabel('year')
plt.ylabel('Total Volume')
plt.title('Volume venduto per anno')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
plt.figure(figsize=(10, 6))
sns.pairplot(df.select_dtypes(include=['number']), hue='year', palette='coolwarm', corner=True)
plt.show()
```

```python
#confronto organic e conventional
dati = df.groupby('type').agg({
    'AveragePrice': 'mean',
    'Total Volume': 'sum'  
}).reset_index()

#prezzo medio
plt.figure(figsize=(8,6))
sns.barplot(hue='type', y='AveragePrice', data=dati, palette="pastel")
plt.title("Confronto Prezzo Medio: Convenzionale vs Organico")
plt.xlabel("type")
plt.ylabel("AveragePrice")
plt.show()

#volume totale
plt.figure(figsize=(8,6))
sns.barplot(hue='type', y='Total Volume', data=dati, palette="muted")
plt.title("Confronto Volume Totale Venduto: Convenzionale vs Organico")
plt.xlabel("type")
plt.ylabel("Total Volume")
plt.show()
```

```python
volume_regione = df_stati.groupby('region')['Total Volume'].sum()

# distribuzione volume totale venduto per regione
plt.figure(figsize=(8,8))
plt.pie(volume_regione, labels=volume_regione.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribuzione del Volume Totale Venduto per città")
plt.axis('equal')
plt.show()
```

```python
regioni_da_escludere = ["TotalUS", "West", "East", "Southwest", "Southeast", "Northeast", "Midsouth", "Plains","SouthCentral", "California"]
# Lista delle regioni ammesse 
regioni_ammissibili = ["East", "West", "Southeast", "North", "Northeast", "Southwaest", "Midwest", "SouthCentral", "South"]

# Creazione del nuovo DF mantenendo solo le regioni nella lista
df_raggruppamentiCardinali = df[df['region'].isin(regioni_ammissibili)]
volume_regione = df_raggruppamentiCardinali.groupby('region')['Total Volume'].sum()

# distribuzione volume totale venduto per regione
plt.figure(figsize=(8,8))
plt.pie(volume_regione, labels=volume_regione.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribuzione del Volume Totale Venduto per punti cardinali(Est, Ovest, Nord-Est, Sud, ecc ecc")
plt.axis('equal')
plt.show()
```

```python
# volume totale per regione
plt.figure(figsize=(12,6))
sns.boxplot(hue='region', y='Total Volume', data=df_stati, palette="vlag",legend=False)
plt.title("Distribuzione del Volume Totale Venduto per Regione")
plt.xlabel("region")
plt.ylabel("Total Volumee")
plt.xticks(rotation=45)
plt.show()

# prezzo medio per regione
plt.figure(figsize=(12,6))
sns.boxplot(hue='region', y='AveragePrice', data=df_stati, palette="coolwarm")
plt.title("Distribuzione del Prezzo Medio per Regione")
plt.xlabel("region")
plt.ylabel("Prezzo Medio")
plt.xticks(rotation=45)
plt.show()
```

```python
vendite_plu_anno = df.groupby('year')[['4046', '4225', '4770']].sum().reset_index()

print(vendite_plu_anno)

# confronto le vendite dei PLU
plt.figure(figsize=(10,6))
bar_width = 0.25
x = vendite_plu_anno['year']

# Spostiamo le barre per evitare sovrapposizioni(mi sono fatto aiutare da chat qua perche non mi appariva nulla e ci sono impazzito)
plt.bar(x - bar_width, vendite_plu_anno['4046'], width=bar_width, label='4046')
plt.bar(x, vendite_plu_anno['4225'], width=bar_width, label='4225')
plt.bar(x + bar_width, vendite_plu_anno['4770'], width=bar_width, label='4770')

plt.xlabel("Anno")
plt.ylabel("Vendite Totali")
plt.title("Confronto Vendite per PLU nel Tempo")
plt.legend()
plt.xticks(x)
plt.show()
```

```python
#prezzo medio per tipo
plt.figure(figsize=(8,6))
sns.boxplot(hue='type', y='AveragePrice', data=df, palette="Set3")
plt.title("Distribuzione del Prezzo Medio per Tipo")
plt.xlabel("Tipo (Convenzionale vs Organico)")
plt.ylabel("Prezzo Medio")
plt.show()
```
